```python
# Path to the folder containing CSV files
folder_path = "/Users/yuninghu/Downloads/Geo_Metadata_Project"

# List to store data frames
data_frames = []

# Iterate through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path, low_memory=False)
        
        # Append the DataFrame to the list
        data_frames.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(data_frames, ignore_index=True)

# Display the combined DataFrame
print(combined_df)
```

```python
df= combined_df 

df
```

```python
df['Sex_merge'] = df['Sex'].fillna(df['sex']).fillna(df['SEX'])
df['Age_merge'] = df['age'].fillna(df['Age']).fillna(df['AGE'])
df['Source_merge'] = df['source'].fillna(df['cell type']).fillna(df['Cell type'])
df['Strain_merge'] = df['strain'].fillna(df['Strain'])
```

```python
dh= df.loc[df['organism'] == 'Homo sapiens']
dn= df.loc[df['organism'] != 'Homo sapiens']
dh
```

```python
dh.loc[:, 'race_1'] = dh['race'].notnull().astype(int)
dh.loc[:, 'Sex_merge_1'] = dh['Sex_merge'].notnull().astype(int)
dh.loc[:, 'Age_merge_1'] = dh['Age_merge'].notnull().astype(int)
dh.loc[:, 'Source_merge_1'] = dh['Source_merge'].notnull().astype(int)
dh.loc[:, 'organism_1'] = dh['organism'].notnull().astype(int)


dh
```
